```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
data=pd.read_csv("D:\Iris.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
ds=data["Species"].to_numpy().reshape(-1,1)
```

```python
ds
```

```python
le.fit_transform(ds)
```

```python
data["Species"]=le.fit_transform(ds)
```

```python
data
```

```python
data["Species"].value_counts()
```

```python
sns.FacetGrid(data,hue="Species",size=7).map(plt.scatter,"PetalLengthCm","PetalWidthCm").add_legend()
```

```python
iris_data=data.iloc[:,:5]
iris_data
```

```python
iris_species=data.iloc[:,5:6]
iris_species
```

```python
model=DecisionTreeClassifier()
#model=DecisionTreeClassifier(criterion="entropy",max_depth=3)
```

```python
model.fit(iris_data,iris_species)
print(model)
```

```python
model.score(iris_data,iris_species)
```

# Make Predictions

```python
expected=iris_species
predicted=model.predict(iris_data)
```

# summarize the fir of the model

```python
print(metrics.classification_report(expected,predicted))
```

```python
print(metrics.confusion_matrix(expected,predicted))
```
